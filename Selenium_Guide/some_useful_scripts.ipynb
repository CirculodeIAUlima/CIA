Classify & Reorder Files in a Folder W Subdirectories

```python
file_types = {
    'Notes': ['txt', 'doc', 'docx', 'odt', 'pdf', 'pptx'],
    'Images': ['jpeg', 'jpg', 'png'],
    'PDFs': ['pdf'],
    'Datasets': ['csv', 'xlsx', 'json'],
    'Videos': ['mp4'],
    'Additional': ['task'] # Add more file types here pe causa
}

def get_folder_name(extension):
    for key, value in file_types.items():
        if extension in value:
            return key
    return None

def move_files(source_directory):
    import os
    for root, dirs, files in os.walk(source_directory):
        for file in files:
            extension = file.split('.')[-1].lower()
            folder_name = get_folder_name(extension)
            if folder_name:
                target_folder = os.path.join(source_directory, folder_name)
                if not os.path.exists(target_folder):
                    os.makedirs(target_folder)
                target_path = os.path.join(target_folder, file)
                # Check if the target file already exists and rename if necessary
                counter = 1
                while os.path.exists(target_path):
                    # Split the file into name and extension
                    file_name, file_extension = os.path.splitext(file)
                    # Add a number to the filename
                    new_file_name = f"{file_name}_{counter}{file_extension}"
                    target_path = os.path.join(target_folder, new_file_name)
                    counter += 1
                # Move the file to the target folder
                os.rename(os.path.join(root, file), target_path)

# Main Execution Block
source_directory = r'C:\Users\Ignacio\Desktop\TESISPRUEBA'
move_files(source_directory)
```
